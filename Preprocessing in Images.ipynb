```python
import os
from os.path import join
import random
from tqdm.auto import tqdm # showing the user how many pages have been processed and how many are remaining. It's a useful tool for visualizing the progress of long-running loops.

from glob import glob      # lib in python to search in the directory for specific type of file like pdf
from pdf2image import convert_from_path
from PIL import Image, ImageEnhance      # dealing with the contrast of the image


data_dir = ""
```

```python
def preprocess_image(image, max_width=600):
  '''
  preprocess image to resuce size and prepare for OCR:
  ARGS : image: PIL Image Object
         max_width: max width in pixels (height auto-calculated to maintain ratio)
  Steps:
         convert to graysclae (color is not important)
         resize to reasonable width while maintaining aspect ratio
         increase constrast (optional)
         RETURN as JPEG with optimization

  '''
  gray_image = image.convert('L')

  if gray_image.width > max_width:
    ratio = max_width / gray_image.width
    new_height = int(gray_image.height * ratio)
    resized_image = gray_image.resize((max_width, new_height), Image.LANCZOS)

  enhancer = ImageEnhance.Contrast(resized_image)
  enhanced_image = enhancer.enhance(1.5)              # to is the value between white and black in image could be possetive or negative
  return enhanced_image


def convert_pdf_to_images(pdf_path, output_base_dir, max_width):

  '''
  convert a PDF file to a set of preprocessed images.

  ARGS:
        pdf_path: path to the PDF file
        output_base_dir: base directory to save images
        max_width: max width in pixels for output images (default: 600)

  '''
  # get pdf filename without extention :

  pdf_name = os.path.splitext(os.path.basename(pdf_path))[0]

  # create output directory :

  output_dir = join(output_base_dir, pdf_name)
  os.makedirs(output_dir, exist_ok=True)

  print(f"converting {pdf_name}.pdf to images ")

  images = convert_from_path(pdf_path, dpi=200) # resolution of the image
  Generated_paths = []

  # process each page :

  for page_number , image in enumerate(images, start=1):

    preprocessed_image = preprocess_image(tqdm(image), max_width)


  # save as JPEG without optimization :

    output_path = join(output_dir, f"page_{page_number:03d}.jpg") # 03d is used for makeing the pattern of the image name is page_001.jpg instead of page_1.jpg
    preprocessed_image.save(output_path, "JPEG", quality=85, optimize=True)

    print(f"Saved Page {page_number} -> {output_path}")

    Generated_paths.append(output_path)

  return Generated_paths
```

```python
sample_pdf_path = f"/content/data/download.pdf"
output_base_dir = f"/content/results"

generated = convert_pdf_to_images(
    pdf_path=sample_pdf_path,
    output_base_dir=output_base_dir,
    max_width = 200)
```
